```python
import csv
import pandas as pd
import ast
import os

not_found = []

def extract_comments(comment_str):
    # Safely convert the string to a Python list using ast.literal_eval
    try:
        comment_list = ast.literal_eval(comment_str)
        # Extract 'comment' values from the list of dictionaries
        comments = [item['comment'] for item in comment_list if 'comment' in item]
        return comments
    except (ValueError, SyntaxError):
        return []

filename = "/u1/mdr614/On the compleness of review comments/Notebook_on the compleness/BERT_CONFU_AUTHOR/test_bert_author.xlsx"
df = pd.read_excel(filename)

for index, row in df.iterrows():
    f=0
    results = []
    if index>=0:
        print(row['Project'])
        comment = row['comment']
        label = row['actual']
        if label==1:
            cls = 'confusing'
        else:
            cls = 'non-confusing'
        
        
        project_filename = '/u1/mdr614/On the compleness of review comments/Notebook_on the compleness/All_Raw/apache_'+row['Project']+'_ChangeRequestComments.csv'
        
        if not os.path.exists(project_filename):
            project_filename = '/u1/mdr614/On the compleness of review comments/Notebook_on the compleness/All_Raw/'+row['Project']+'_ChangeRequestComments.csv'
        
        df_project = pd.read_csv(project_filename)
               
        for index1, row1 in df_project.iterrows():
            comment_str = row1['Comment']
            # Extract comments from the current row
            comments_list = extract_comments(comment_str)
            
            
            
            if comment in comments_list:
                #print(comment_str)
                f = 1
                results.append({
                'CCR': comments_list[0],
                'response': comment,
                'label': cls
            })
                #print(results)
                result_df = pd.DataFrame(results)
                file_path = 'RQ4_dataset_more.csv'
                result_df.to_csv(file_path, mode='a', index=False, header=not os.path.exists(file_path))  
                break

        if f==0:
            print("Index: ",index," No found")
            not_found.append(index)

            
```
