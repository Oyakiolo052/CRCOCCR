```python
import pandas as pd
import requests

# Path to your CSV file
file_path = 'random_sample_noAuthor_reply.csv'

# Load the CSV file into a DataFrame
data_df = pd.read_csv(file_path)

# Display the DataFrame to verify it's loaded
print("DataFrame loaded:")
#print(data_df)  # Uncomment this line to display the DataFrame

# Iterate through each row in the DataFrame
for index, row in data_df.iterrows():
    if index>=9921:
        print("index: ",index)
        project = row['Project']
        pull = row['pull']



        print("Project:", project)
        print("Pull Number:", pull)

        # Construct the API URL
        api_url = f'https://api.github.com/repos/apache/{project}/pulls/{pull}'
        print("API URL:", api_url)

        # Make the API request
        response = requests.get(api_url, auth=(username, token))
        data_json = response.json()  # Get the JSON response
        #print(data_json)


        # Check if the JSON response is empty or a dictionary
        if not data_json:  # Check for empty or dict response
            print("Json file empty!")
        else:
            print("Merged status:", data_json.get('merged'))  # Get 'merged' status, default to 'N/A'

            result = {
                    'project': project,
                    'pullNumber': pull,
                    'CCR': row['CCR'],
                    'author': row['Author'],
                    'merged': data_json.get('merged')
                }
            pd.DataFrame([result]).to_csv('noAuthorReply_accept_reject_result.csv', mode='a', header=not pd.io.common.file_exists('noAuthorReply_accept_reject_result.csv'), index=False)


```
